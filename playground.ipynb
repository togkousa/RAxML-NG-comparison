```python
import pandas as pd
```

```python
df=  pd.read_parquet("results/027.phy_GTR+G/search/m1-raxmlng/m1-raxmlng.results.trees.parquet")
df["plausible"] = [True] * 10 + [False] * 10
df.to_parquet("results/027.phy_GTR+G/search/m1-raxmlng/m1-raxmlng.results.trees.parquet")
```

```python
from plotly import graph_objects as go



fig = go.Figure(
    go.Bar(
        x=df.plausible.value_counts().index,
        y=df.plausible.value_counts().values,
        name="m1"
    )
)

fig.add_trace(
    go.Bar(
        x=df.plausible.value_counts().index,
        y=df.plausible.value_counts().values,
        name="m2"
    )
)
```
